```python
## DATAFRAME VIEW OF FOLDER

import os, sys
cwd = os.getcwd()
cwd_files = os.listdir()
is_dir = [os.path.isdir(x) for x in cwd_files]
is_file = [os.path.isfile(x) for x in cwd_files]

import pandas as pd
df_dir = pd.DataFrame({'Item':cwd_files,'is_dir':is_dir,'is_file':is_file})

dir_idx = df_dir[df_dir['is_dir']==True].index
file_idx = df_dir[df_dir['is_file']==True].index

df_dir.loc[dir_idx,'type'] = 'Folder'
df_dir.loc[file_idx,'type'] = 'File'

df_dir.drop(['is_dir','is_file'],axis=1,inplace=True)


folders = df_dir.groupby('type').get_group('Folder').sort_values('Item')#sort_values('Items',ascending=True)
files = df_dir.groupby('type').get_group('File').sort_values('Item')
df_directory = pd.concat([folders,files],axis=0)
df_directory.reset_index(drop=True)
```
